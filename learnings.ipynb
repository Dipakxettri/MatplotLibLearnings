```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Simple line plot

```python
# Data 
x = [1,2,3,4]
y = [1,2,10,44]

# lets plot it 
plt.plot(x,y)
plt.grid() # showing grid also
plt.show()
```

# Adding Titles & Labels

```python
plt.title("Data plot")
plt.ylabel("x values")
plt.xlabel("y values")
plt.show
```

# Changing Line Style & Color

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
# Common options:
# Option	Values
# color	'red', 'blue', 'green', '#ff8800'
# linestyle	'-' (solid), '--' (dashed), ':' (dots)
# marker	'o', 's', '*', 'x'
```

```python
# Univariate - Numerical
```

```python
data = {
    "salery" : [222,444,2364,363,383,383,3883,2991,1001]
}

df = pd.DataFrame(data)
df.shape
```

# Line Plot

```python
# Line Plot
plt.plot(df["salery"])

# some styling or customization
plt.plot(df["salery"], color="red", marker="o", linestyle=":", linewidth ="2")
plt.grid()
plt.show
```

# Histogram

```python
# plt.hist(df["salery"])
plt.hist(df["salery"], color="green")
plt.show
```
