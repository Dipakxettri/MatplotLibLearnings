```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Simple line plot

```python
# Data 
x = [1,2,3,4]
y = [1,2,10,44]

# lets plot it 
plt.plot(x,y)
plt.grid() # showing grid also
plt.show()
```

# Adding Titles & Labels

```python
plt.title("Data plot")
plt.ylabel("x values")
plt.xlabel("y values")
plt.show
```

# Changing Line Style & Color

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
# Common options:
# Option	Values
# color	'red', 'blue', 'green', '#ff8800'
# linestyle	'-' (solid), '--' (dashed), ':' (dots)
# marker	'o', 's', '*', 'x'
```

```python
# matplot lib can be implimented using 2 types:
# Pyplot API and object oriented API

# Pyplot API :
```

```python
# Univariate - Numerical
```

```python
data = {
    "salery" : [222,444,2364,363,383,383,3883,2991,1001]
}

df = pd.DataFrame(data)
df.shape
```

# Line Plot

```python
# Line Plot
plt.plot(df["salery"])

# some styling or customization
plt.plot(df["salery"], color="red", marker="o", linestyle=":", linewidth ="2")
plt.grid()
plt.show
```

# Histogram

```python
# plt.hist(df["salery"])
plt.hist(df["salery"], color="green")
plt.show
```

# Boxplot

```python
plt.boxplot(df["salery"])
plt.show

# box plot has two end points one repersents a max and one min values
# min is connected to the lower quartile known as (Q1) that represnets 25 % of the data so 25 % of data is below this point and that line is called whisker
# mean is between median and lower quartile
# median(Q2) is in between of a box  
# Upper quartile (Q3) represnets 75 % of a data is bellow
# Q1 to Q3 is known as interquartile range (IQR)
# outliers are either greather or lower then min and max, they are extrime points
```

```python
df["dept"] = ["HR","Support","Head","Finance","HR","Support","Head","Finance","HR"]
df.info()
```

```python
# we can see from the flowing output of 'df.info()' that our new updated data frame has object column as a type
# so we can't plot them using line, histogram and boxplot insted we use piechart
```

# PieChart

```python
count = df["dept"].value_counts() # count unique number of values
count
```

```python
# plotting pie chart
plt.pie(count, labels=count.index, autopct="%1.1f", explode = [0,0.0,0,0.2])# explode helps to make any piece seperate
plt.axes
plt.show()
```
