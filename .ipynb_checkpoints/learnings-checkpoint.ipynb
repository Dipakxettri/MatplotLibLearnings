```python
import matplotlib.pyplot as plt
```

# Simple line plot

```python
# Data 
x = [1,2,3,4]
y = [1,2,10,44]

# lets plot it 
plt.plot(x,y)
plt.show()
```

# Adding Titles & Labels

```python
plt.title("Data plot")
plt.ylabel("x values")
plt.xlabel("y values")
```

# Changing Line Style & Color

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
# Common options:
# Option	Values
# color	'red', 'blue', 'green', '#ff8800'
# linestyle	'-' (solid), '--' (dashed), ':' (dots)
# marker	'o', 's', '*', 'x'
```
